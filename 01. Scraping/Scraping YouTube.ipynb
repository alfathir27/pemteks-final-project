## **Data Scraping**

```python
def get_video_comments(youtube, video_id, limit=500):
    all_comments = []
    count = 0

    # Ambil judul video untuk identifikasi
    video_request = youtube.videos().list(
        part="snippet",
        id=video_id
    )
    video_info = video_request.execute()
    video_title = video_info["items"][0]["snippet"]["title"]

    request = youtube.commentThreads().list(
        part="snippet",
        videoId=video_id,
        maxResults=100,
        textFormat="plainText",
        order='relevance'
    )

    while request and count < limit:
        response = request.execute()

        for item in response["items"]:
            if count >= limit:
                break

            snippet = item["snippet"]["topLevelComment"]["snippet"]

            comment_info = {
                "video_id": video_id,
                "video_title": video_title,
                "penulis": snippet["authorDisplayName"],
                "teks_komentar": snippet["textDisplay"],
                "jumlah_like": snippet["likeCount"],
                "tanggal_publish": snippet["publishedAt"],
            }

            all_comments.append(comment_info)
            count += 1

        request = youtube.commentThreads().list_next(request, response)

    print(f"‚úî {video_id} ‚Äî {len(all_comments)} komentar diambil")
    return all_comments
```
